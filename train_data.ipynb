```python
# Install kagglehub if not already installed
!pip install kagglehub[pandas-datasets]

import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the exact filename from the dataset
file_path = "train.csv"
# Load dataset into a Pandas DataFrame
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "abhi8923shriv/sentiment-analysis-dataset",
  file_path,
  pandas_kwargs={"encoding": "unicode_escape"}  # Add encoding parameter
)

# Display the first 5 records
print("First 5 records:", df.head())
```

```python
df = df.dropna()
print(df.shape)  # Should ideally be (27481, 10)
```

```python
train_data=df[['text','sentiment']]
train_data.head()
```

```python
train_data.describe()
```

```python
train_data.isna().any()
```

```python
Unique_columns=train_data['sentiment'].unique()
print(Unique_columns)
```

```python
dict={Unique_columns[i]: i for i in range(3)}
print(dict)
train_data.replace(dict,inplace=True)
train_data.head()
```

```python
#implementing bag of words
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
vectorizer = CountVectorizer()  # Create the BoW model
X_bow = vectorizer.fit_transform(train_data['text'])  # Fit and transform the text column
# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Convert sparse matrix to DataFrame
bow_df = pd.DataFrame(X_bow.toarray(), columns=feature_names)

# Display the BoW representation
print(bow_df)
```
