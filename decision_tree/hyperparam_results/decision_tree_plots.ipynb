```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the saved CSV
df = pd.read_csv('decision_tree_comparison.csv')

# Set plot style
sns.set(style="whitegrid")

# ----------- 1. Plot: Accuracy vs max_depth -------------
plt.figure(figsize=(10, 6))
grouped = df.groupby('max_depth')[['accuracy_bow', 'accuracy_tfidf']].mean().reset_index()
sns.lineplot(x='max_depth', y='accuracy_bow', data=grouped, marker='o', label='BoW Accuracy')
sns.lineplot(x='max_depth', y='accuracy_tfidf', data=grouped, marker='o', label='TF-IDF Accuracy')
plt.title("Accuracy vs max_depth")
plt.ylabel("Accuracy")
plt.xlabel("Max Depth")
plt.legend()
plt.tight_layout()
plt.show()

# ----------- 2. Plot: Accuracy vs min_samples_split -------------
plt.figure(figsize=(10, 6))
grouped = df.groupby('min_samples_split')[['accuracy_bow', 'accuracy_tfidf']].mean().reset_index()
sns.lineplot(x='min_samples_split', y='accuracy_bow', data=grouped, marker='o', label='BoW Accuracy')
sns.lineplot(x='min_samples_split', y='accuracy_tfidf', data=grouped, marker='o', label='TF-IDF Accuracy')
plt.title("Accuracy vs min_samples_split")
plt.ylabel("Accuracy")
plt.xlabel("min_samples_split")
plt.legend()
plt.tight_layout()
plt.show()

# ----------- 3. Plot: Accuracy vs min_samples_leaf -------------
plt.figure(figsize=(10, 6))
grouped = df.groupby('min_samples_leaf')[['accuracy_bow', 'accuracy_tfidf']].mean().reset_index()
sns.lineplot(x='min_samples_leaf', y='accuracy_bow', data=grouped, marker='o', label='BoW Accuracy')
sns.lineplot(x='min_samples_leaf', y='accuracy_tfidf', data=grouped, marker='o', label='TF-IDF Accuracy')
plt.title("Accuracy vs min_samples_leaf")
plt.ylabel("Accuracy")
plt.xlabel("min_samples_leaf")
plt.legend()
plt.tight_layout()
plt.show()

# ----------- 4. Plot: Accuracy vs criterion (bar plot) -------------
plt.figure(figsize=(8, 6))
grouped = df.groupby('criterion')[['accuracy_bow', 'accuracy_tfidf']].mean().reset_index()
grouped.plot(kind='bar', x='criterion', y=['accuracy_bow', 'accuracy_tfidf'], figsize=(8, 6))
plt.title("Accuracy vs criterion")
plt.ylabel("Accuracy")
plt.xlabel("Criterion")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
